Using markdown notation you can do cells with text. I already made the notebook to be a python notebook.

This one is just a basic example, in case you want to use it. I think it can be an easy and fast way to present code to them.

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing a seed for reproducibility
np.random.seed(10)

mu, sigma = 0, 0.1 # mean and standard deviation
x = np.random.normal(mu, sigma, 1000)

# This is a pretty bad way to do the sum, but just as an example of a for loop
y = np.empty_like(x)
y[0] = x[0]
for i in range(999):
    y[i+1] = x[i+1] + x[i]



plt.figure(figsize=(15,4))
plt.plot(np.arange(0, 1000, 1), x, label='X')
plt.plot(np.arange(0, 1000, 1), y, label='Y')
plt.xlabel('realization')
plt.ylabel('value')
plt.legend()
plt.show()
    
```
